#Basic Operations

Now that you know how to mount your drive, let's take it a step further and see how to do basic operations in colab.


To **print the current working directory** run:
```
pwd
```
To **change the directory** run (replace *Destination* with a valid path e.G. /content/drive/MyDrive/AiXDesignKitchen/Workshop) :

```
cd Destination
```
To **create a new folder** run (replace *foldername* with a name of your choice:

```
mkdir foldername
```
To **create a file** you can run the following function:
```
open("Filename.xyz", "x")
```

To **delete a file** you can run this command:
```
rm filepath/filename
```
To **Upload a file** from your Local Machine to your drive run:
```
from google.colab import files

filename = list(files.upload().keys())[0]
```

To comment use **#**
```
 #This is a comment
```
