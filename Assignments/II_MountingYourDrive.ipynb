# Mounting Google Drive

In Colab there are two ways to mount your Google Drive.







**Method A:** Using the Interface

This method is most useful when you work on a project that you create yourself from scratch.

--- 

1. Click on the file symbol on the left side.


Colab_Files_01.png





2. Click on the mount drive symbol

Colab_Files_02.png

3. Confirm
Colab_Files_03.png

Now your drive is mounted and listed as 'drive' in the file browser. To unmount press the unmount drive symbol.


**Method B:** Using Code

This method is most useful when you cloned someone elses project and continue your work from there.

1. Add the following to a code block and run it:
```
from google.colab import drive
drive.mount('/content/drive')
```
