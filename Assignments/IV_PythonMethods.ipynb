In python there is no one size fits all solution. There are many different approaches of achieving the same results utilizing different methods and modules. In the following example we're going to learn how to import various functions to ultimately be able to display an image in our notebook.

<h3>Method #1</h3>

In this method we try to install the package **opencv-python** to use **cv2** functions.

```python
!pip install opencv-python
```

```python
import cv2
img = cv2.imread('logo.png')
#img = cv2.imread('logo.png', cv2.IMREAD_UNCHANGED) # cv2.IMREAD_UNCHANGED fixes display errors
cv2_imshow(img)
```

Since **opencv-python** is not supported in **Jupyter Notebooks** we use a patch provided by Google Colab

```python
from google.colab.patches import cv2_imshow # This fixes incompatibility issues with opencv-python on Jupyter notebooks
```

<h3>Method #2</h3>

```python
from IPython.display import display, Image
display(Image(filename='logo.png'))
#display(Image(filename='logo.png', width=400))
```

<h3>Method #3</h3>

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('logo.png')
imgplot = plt.imshow(img)
plt.show()
```
